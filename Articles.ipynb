```python
def newART(show=True):
    def make_clickable(val):
        return '<a target="_blank" href="{}">Link</a>'.format(val)
    from IPython.display import clear_output
    df = pd.read_csv('Articles.csv',index_col=0)
    add = 0
    n = df.index[-1]+1
    while add == 0:
        print('Article no. {}'.format(n))
        title = input('Title: ')
        link = input('Link to article: ')
        method = input("Method ['BEH', 'DTI', 'DTI+fMRI', 'SBM', 'VBM', 'fMRI', 'fMRI+SBM'] ")
        authors = input('Authors: ')
        year = int(input('Year: '))
        inout = int(input('In (1) or Out (0): '))
        rsn = input('Reason: ')
        clear_output(wait=True)
        tempDF = pd.DataFrame({
        'Authors':authors,
        'Year':year,
        'Title':title,
        'Method':method,
        'Link':make_clickable('http://sci-hub.tw/{}'.format(link)),
        'In/Out':inout,
        'Reason':rsn},index=[0])[['Authors','Year','Title','Method','Link','In/Out','Reason']]
        df = pd.concat([df,tempDF]).reset_index(drop=True)
        df.index = range(1,len(df)+1,1)
        cnt = int(input('Continue (1) or Break (0)? '))
        n += 1
        if cnt == 0:
            break
    df.to_csv('Articles.csv')
    print('*****DONE*****')
    if show==True:
        HTML(df.to_html(escape=False))
    return df
```
